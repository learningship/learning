```python
import numpy as np
import matplotlib.pyplot as plt
f1=0.1
f2=0.3
N=64
r=[0 for i in range(N)]
n=np.arange(0,N,1)
y=np.sin(2*np.pi*f1*n+np.pi/3)+np.sin(2*np.pi*f2*n+np.pi/4)
plt.figure(1)
for i in range(N):
    plt.plot([i,i],[0,y[i]])
plt.show()
plt.figure(2)
Y1=np.fft.fft(y)
for i in range(N):
    plt.plot([i,i],[0,np.abs(Y1[i])])
plt.show()
for j in range(N):
    for i in range(N-j):
        r[j]=r[j]+y[i]*y[i+j]
Y2=np.fft.fft(r)
plt.figure(3)
for i in range(N):
    plt.plot([i,i],[0,Y2[i]])
plt.show()
```
