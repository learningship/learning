```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
f1=0.2
f2=0.201
N=128
k=20
M=2
n=np.arange(0,k)
Rx=[]
p=[0 for i in range(N)]
f=[]
TRx=np.zeros((k,k))
a=[]
x=10*np.sin(2*np.pi*f1*n)+10*np.cos(2*np.pi*f2*n)+np.random.randn(k)
for j in range(k):
    Rx.append(np.sum(x[j:k]*x[0:k-j]))
#print(np.append(Rx[2:0:-1],Rx[1:5]))
for i in range(k):
    TRx[i]=np.append(Rx[i::-1],Rx[1:k-i])
plt.figure(1)
plt.plot(np.fft.fft(Rx))
plt.plot(np.square(np.abs(np.fft.fft(x))))
plt.show()
w,v=LA.eig(TRx)
e=v[:,4:k]
#w=np.abs(np.diag(w))
print(TRx[0],TRx[1],'w:',np.abs(w),v.shape,e.shape)
i=np.arange(k)
for dw in range(N):
        a=np.exp(1j*i*2*np.pi*dw/N)
        p[dw]=1/(a.T@e@e.T@a)
#print(a.T.shape,dw.shape)
plt.figure(2)
plt.plot(np.arange(N)/N,np.abs(p))
```
